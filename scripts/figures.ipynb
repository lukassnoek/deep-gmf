# Figures for *deep-GMF* project

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

## Data import

```python
model = 'ResNet10'
dataset = 'gmf'
targets = ['id', 'id+gender+ethn', 'id+gender+ethn+l+xt+yt+zt+xr+yr+zr']

dfs = [pd.read_csv(f'../results/{model}_dataset-{dataset}_target-{target}.tsv', sep='\t')
       for target in targets]
histories = [pd.read_csv(f'../models/{model}_dataset-{dataset}_target-{target}_history.csv')
             for target in targets]
for h in histories:
    if 'val_id_loss' in h:
        plt.plot(h['val_id_loss'])
    else:
        plt.plot(h['val_loss'])
    
plt.legend(targets)
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
ax = sns.lineplot(x='op_nr', y='corr', hue='factor', data=dfs[0], ax=ax)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1.05, 1))
sns.despine()

ax.grid(ls='--', lw=0.35)
ax.set_ylim(-.05, 1)
ax.set_xlim(-.05, dfs[0]['op_nr'].max() + 1)
```
